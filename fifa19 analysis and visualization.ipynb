```python
# Convert values to float
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0
```

```python
# Defenders
position = df['Position'].unique()

# groupby positions
gk_pos = ['GK']
def_pos = ['CB','LCB','RCB','LWB', 'LB', 'RB', 'RWB']
mid_pos = ['RCM','LCM','CAM','CDM','RM','LAM','RDM','RAM','CM','LM','LDM']
wingers = ['LW','RW']
fwd_pos = ['RF','LF','ST','RS','CF','LS']
```

```python
# Standard practice, DRY
def best_overall_by_pos(position,benchmark=85):    
#     put the positions in a dict
    pos_dict = {
        'goalkeepers':gk_pos,
        'defenders':def_pos,
        'midfielders':mid_pos,
        'forwards':fwd_pos,
        'wingers':wingers
    }
    
#     get the players positions
    players = df[df['Position'].isin(pos_dict.get(position))]

#     overall best by benchmark
    if benchmark >= 85:
        best_players = players[players['Overall'] > benchmark]
    else:
        print('C\'mon! That\'s too low for top players')
        return None
    return best_players
    
#     plot the graph
def plot_graph(best_players):    
    HOVERTOOLTIPS = HoverTool(tooltips=[
                                    ('Name', '@Name'),
                                    ('Nationality', '@Nationality'),
                                    ('Club', '@Club'),
                                    ('Age', '@Age')
                                   ]
                          )
    title = best_players['Position'].all()
    
    plot = figure(title=f'Best {title} in fifa19',
                  x_axis_label='Value',
                  y_axis_label='Overall',
                  tools=[HOVERTOOLTIPS])
    plot.circle('Value', 'Overall', size=10, source=best_players)
    show(plot)
```

```python
df2 = best_overall_by_pos('midfielders',85)
df2.head(5)
```

```python
plot_graph(df2)
```
